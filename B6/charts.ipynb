```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cleaned_Spotify_2024_Global_Streaming_Data.csv")
df.isna().sum()
```

```python
df.tail()
```

```python
# vẽ biểu đồ đường cho monthly Listeners (millions)
df.plot()
```

# BIỂU ĐỒ PLOT
- biểu đồ đường cong

```python
# ve bieu do duong cho Monthly Listeners (Millions) 
plt.figure(figsize=(14,6))
plt.plot(df["Monthly Listeners (Millions)"], marker='o', linestyle='-', linewidth=2, color='black')
plt.ylabel(df["Monthly Listeners (Millions)"], fontsize=14)
plt.title("Biểu đồ đường số người nghe theo tháng theo từng album", fontsize=16) 

plt.grid(True, linestyle='--', alpha=0.5)

# hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

# BAR
- biểu đồ cột

```python
# gom nhóm data
artist_streams = df.groupby("Artist")["Total Streams (Millions)"].sum().nlargest(10)
artist_streams
```

```python
# ve bieu do cot ngang 
plt.figure(figsize=(10, 15)) 
artist_streams.plot(kind="barh", color="#111", edgecolor="skyblue")

plt.title("Top 10 nghệ sĩ có stream cao nhất", fontsize=14) 
plt.xlabel("Total streams (mi)", fontsize=12) 
plt.ylabel("Artist", fontsize=12)

# dao truc de nghe si cao nhat nam o tren 
plt.gca().invert_yaxis()

plt.grid(axis="x", linestyle="--", alpha=0.5) 
plt.tight_layout()

plt.show ()
```
