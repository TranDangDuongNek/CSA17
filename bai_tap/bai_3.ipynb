```python
import pandas as pd
data = pd.read_csv("hw3_data.csv")

# Có bao nhiêu người trên 30 tuổi
over_30 = data[data["Age"] > 30]
print("Câu 1: Số người trên 30 tuổi là:", len(over_30))

# tổng số tiền lương của tất cả các người trong dữ liệu
total_salary = data["Salary"].sum()
print("Câu 2: Tổng số tiền lương là:", total_salary)

# Có bao nhiêu người có lương > 50000 và tuổi < 40
filtered = data[(data["Salary"] > 50000) & (data["Age"] < 40)]
print("Câu 3: Số người có lương lớn hơn 50000 và tuổi dưới 40 là:", len(filtered))
```
